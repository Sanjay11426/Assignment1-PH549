```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand

# Number of Steps
num_steps = 1000

# Number of Walkers
num_walks = 100

# probability
w = 0.9

# Variables for Mean Square distance Calculation
x2 = np.zeros(num_steps) 
xx = np.zeros(num_steps)

# Variable for storing step length at each time
step_x = np.zeros(num_steps)

# Variable for time steps
t = np.zeros(num_steps)
for i in range (num_steps):
    t[i] = i

# Variable for walker position at time t
x = np.zeros(num_steps)

for k in range (num_walks):    
    prob_start = np.random.uniform(0,1)
    if prob_start < 0.5:
       step_x[0] = 1
       x[1] = 1
    else: 
       step_x[0] = -1
       x[1] = - 1
    for i in range(2, num_steps):
        prob = np.random.uniform(0,1)
        if prob < w:
            x[i] = x[i - 1] + step_x[i-2]
            step_x[i-1] = step_x[i-2]
        else:
            x[i] = x[i - 1] - step_x[i-2]
            step_x[i-1] = - step_x[i-2]
    for j in range(1,num_steps):
      c = 0
      for l in range(j,num_steps):
        c = c + np.square((x[l] - x[l-j]))
      x2[j] = x2[j] + c/(num_steps-j)
      #  xx[i]= x[i]**2
       # x2[i]= x2[i] + xx[i]

# Average of Displacement Squares
x2 = x2/num_walks
logx2 = np.log(x2[2:])
logt = np.log(t[2:])

# plotting stuff: 
plt.title("1D Random Walk with one step memory")
plt.plot(t, x2)
plt.xlabel('time')
plt.ylabel('Mean Square displacement')
plt.savefig('Part(b).png')
plt.show()
plt.title("1D Random Walk with one step memory")
plt.plot(logt, logx2)
plt.xlabel('Log(t)')
plt.ylabel('Log(MSD')
plt.show()
slope_intercept = np.polyfit(logt,logx2,1)
print(slope_intercept)

```
